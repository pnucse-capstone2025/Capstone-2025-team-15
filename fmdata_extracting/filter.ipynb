```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.patches as patches
import csv as csv
```

```python
path = "./results"
home = path + "/home.csv"
away = path + "/away.csv"
df_home = pd.read_csv(home)
df_away = pd.read_csv(away)
df_away
```

```python
from scipy.signal import savgol_filter

# 예시: 각 좌표 컬럼에 필터 적용
for col in df_home.columns:
    if "_x" in col or "_y" in col:
        df_home[col] = savgol_filter(df_home[col], window_length=25, polyorder=2)

for col in df_away.columns:
    if "_x" in col or "_y" in col:
        df_away[col] = savgol_filter(df_away[col], window_length=25, polyorder=2)
```

```python
df_home.to_csv('./filtered/home.csv', index=False)
df_away.to_csv('./filtered/away.csv', index=False)
```
